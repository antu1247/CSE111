```python
#1
class Student:
    ID =0
    def __init__(self, n,d,a,c):
        self.__Name = n
        self.__Dept = d
        self.__Age = a
        self.cgpa = c
        Student.ID +=1
        
        
    def get_details(self):
        print("ID:",Student.ID)
        print("Name:",self.__Name)
        print("Department:",self.__Dept)
        print("Age:",self.__Age)
        print("CGPA:",self.cgpa)
        
        
    @classmethod
    def from_String(cls, given_string): 
        n,d,a,c= given_string.split("-")
        obj = cls(n,d,a,c)
        return obj
#===================================================================================    
s1 = Student("Samin","CSE",21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE",21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura","EEE",22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details() 
print("-----------------------")
```

```python
class Dates:
    def __init__(self, x):
        self.__x = x
    def getDate(self):
        return self.__x
   
    @classmethod
    def toDashDate(cls, other):
        other= other.replace("/",'-')
        
        obj = cls(other)
        return obj
        
date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate() ):
   print("Equal")
else:
   print("Unequal")
print('It prints equal because here I use class method and in this class method I use replace function to remove dash to hyphene. So both date become same in here and condition matches and prints equal.')
```

```python
#3
class Passenger:
    count = 0
    def __init__(self,name):
        self.name = name
        self.weight = 0
        Passenger.count+=1
        
    def set_bag_weight(self,weight):
        self.weight = weight 
        
    def printDetail(self):
        print("Name:",self.name)
        if self.weight <=20:
            fare = 450
        elif self.weight>=21 and self.weight<=50:
            fare = 450+50
        elif self.weight>50:
            fare = 450+100
        print("Bus Fare:",fare,"taka")
        
print("Total Passenger:", Passenger.count) 
#===============================================================================================
p1 = Passenger("Jack") 
p1.set_bag_weight(90) 
p2 = Passenger("Carol")
p2.set_bag_weight(10) 
p3 = Passenger("Mike") 
p3.set_bag_weight(25) 
print("=========================") 
p1.printDetail() 
print("=========================") 
p2.printDetail() 
print("=========================") 
p3.printDetail() 
print("=========================") 
print("Total Passenger:", Passenger.count)
```

```python
#5
class Fruit:
    fruitCount = 0
    def __init__(self,name,quant):
        self.name = name
        self.count = quant 
        Fruit.fruitCount+=quant
    @classmethod
    def saySomethingGood(cls):
        print("Fruits are good for health")
    @classmethod
    def resetcount(cls):
        Fruit.fruitCount=0
#=============================================================        
        
apples = Fruit("Apple", 3);
pears = Fruit("Pear", 4);
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```python
class Cat:
    Number_of_cats = 0
    def __init__(self,color,action):
        self.color = color
        self.action = action
        Cat.Number_of_cats+=1
     
    @classmethod
    def no_parameter(cls):
        return Cat("White","sitting")
    @classmethod
    def first_parameter(cls,color):
        return Cat(color,'sitting')
    @classmethod
    def second_parameter(cls,action):
        return Cat('Grey',action)
    def printCat(self):
        print(self.color,'cat is',self.action)
    def changeColor(self,color):
        self.color = color
#====================================================================        

print("Total number of cats:", Cat.Number_of_cats) 
c1 = Cat.no_parameter() 
c2 = Cat.first_parameter("Black") 
c3 = Cat("Brown", "jumping") 
c4 = Cat("Red", "purring") 
c5 = Cat.second_parameter("playing")
print("=======================") 
c1.printCat() 
c2.printCat() 
c3.printCat() 
c4.printCat() 
c5.printCat() 
c1.changeColor("Blue") 
c3.changeColor("Purple") 
c1.printCat() 
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self,rad,height):
        self.radius = rad
        self.height=height
        print("Defualt radius=",Cylinder.radius,"and height=",Cylinder.height)
        print("Updated: radius=",self.radius,"and height=",self.height)
        Cylinder.radius = self.radius
        Cylinder.height = self.height
    
    @classmethod    
    def area(cls,rad,height):
        area = (2*math.pi*rad*rad)+(2*math.pi*rad*height)
        print("Area:",area)
    
    # @classmethod    
    # def area(cls,rad,height):
    #     area = (2*math.pi*rad*rad)+(2*math.pi*rad*height)
    #     print("Area:",area)
    
    @classmethod    
    def volume(cls,rad,height):
        volume = math.pi*rad*rad*height
        print("Volume:",volume)
        
    @classmethod
    def swap(cls,rad,height):
        return Cylinder(height,rad)
    
    @classmethod 
    def changeFormat(cls,string):
        lst = string.split('-')
        return Cylinder(int(lst[0]),int(lst[1]))
#=====================================================================    
        
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
